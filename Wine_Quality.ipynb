```python
#Replacing null values with mean

df['fixed acidity'] = df['fixed acidity'].fillna(df['fixed acidity'].mean())
df['volatile acidity'] = df['volatile acidity'].fillna(df['volatile acidity'].mean())
df['citric acid'] = df['citric acid'].fillna(df['citric acid'].mean())
df['residual sugar'] = df['residual sugar'].fillna(df['residual sugar'].mean())
df['chlorides'] = df['chlorides'].fillna(df['chlorides'].mean())
df['pH'] = df['pH'].fillna(df['pH'].mean())
df['sulphates'] = df['sulphates'].fillna(df['sulphates'].mean())
```

```python
df.isnull().sum()
```

```python
fig, arr = plt.subplots(6,2, figsize = (10,8), dpi = 300)
sns.boxplot(x= 'type', y= 'fixed acidity', hue = 'type', data = df, ax = arr[0][0])
sns.boxplot(x= 'type', y= 'volatile acidity', hue = 'type', data = df, ax = arr[0][1])
sns.boxplot(x= 'type', y= 'citric acid', hue = 'type', data = df, ax = arr[1][0])
sns.boxplot(x= 'type', y= 'residual sugar', hue = 'type', data = df, ax = arr[1][1])
sns.boxplot(x= 'type', y= 'chlorides', hue = 'type', data = df, ax = arr[2][0])
sns.boxplot(x= 'type', y= 'free sulfur dioxide', hue = 'type', data = df, ax = arr[2][1])
sns.boxplot(x= 'type', y= 'total sulfur dioxide', hue = 'type', data = df, ax = arr[3][0])
sns.boxplot(x= 'type', y= 'density', hue = 'type', data = df, ax = arr[3][1])
sns.boxplot(x= 'type', y= 'pH', hue = 'type', data = df, ax = arr[4][0])
sns.boxplot(x= 'type', y= 'sulphates', hue = 'type', data = df, ax = arr[4][1])
sns.boxplot(x= 'type', y= 'alcohol', hue = 'type', data = df, ax = arr[5][0])
sns.boxplot(x= 'type', y= 'quality', hue = 'type', data = df, ax = arr[5][1])
```

```python
# Exploratory Data Analysis

fig , ax = plt.subplots(ncols= 6, nrows = 2,figsize=(20,10))
index = 0
ax = ax.flatten()
for col,values in df.items():
    if col!= 'type':
        sns.boxplot(y= col,data = df, ax = ax[index])
        index = index+1
```

```python
fig , ax = plt.subplots(ncols= 6, nrows = 2,figsize=(20,10))
index = 0
ax = ax.flatten()
for col,values in df.items():
    if col!= 'type':
        sns.distplot(values, ax = ax[index])
        index = index+1
```
